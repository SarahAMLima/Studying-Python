**Anonymous functions**

An anonymous function in Python is a function without a name. It can be immediately invoked or stored in a variable. Anonymous functions in Python are also known as lambda functions - Source: FreeCodeCamp


*Basic structure*


x = lambda argumentos: expressao


It can't be used in complex calculous.

```python
soma = lambda x,y: x+y

print (soma(2,5))
```

In geoprocessing, anonymous functions are highly used in GEE associated with another function to read a piece of a datafile, or in other words, to do a clip.


Now, let's do a more practical example, we will calculate the means between two bands to visualize an image.

```python
!pip install spectral
from spectral import imshow
import tifffile as tif
```

```python
img = tif.imread('/content/drive/MyDrive/Python/PDI com Python/Módulo 2/PDicPython - Imagens/L71221071_07120010720_DN.tif')
```

```python
blue = img[:,:,0]
red = img[:,:,2]
```

```python
imshow(blue)
imshow(red)
```

```python
mean = lambda bx,by: (bx + by) / 2
m = mean(blue,red)
imshow(m)
```

What if we want the visible spectre for our image?

We will calculate the mean between the bands blue, green and red.

```python
blue = img[:,:,0]
green = img[:,:,1]
red = img[:,:,2]
mean = lambda bx,by,bz: (bx + by + bz) / 3
m = mean(blue, green, red)
imshow(m, figsize=(12,8))
```
