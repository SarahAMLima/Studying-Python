```python
#importando a geopandas, apelido gpd para evitar de escrever geopandas o tempo todo
import geopandas as gpd
import pandas as pd
```

```python
#Para ler um arquivo: geopandas.read_file() mas nomeamos geopandas como gpd
#gpd.read_file(caminho_arquivo) e caminho_arquivo é o caminho do arquivo no nosso note e ele precisa ser especificado antes de rodarmos a função
#E podemos atribuir uma variável para armazenar todos esses dados, a 'dados_cont'
caminho_arquivo=r'C:\Users\Sarah Lima\Desktop\Geodev\3P\Pygeodev\Semana1\VWM_AREAS_CONTAMINADAS_GEODADOS_CETESB_PTOPoint.shp'
dados_cont=gpd.read_file(caminho_arquivo)
```

```python
#Para conhecer os nossos dados vamos primeiro ver eles: 
dados_cont
```

```python
#Para conhecer os tipo de dados que temos:
dados_cont.dtypes
```

```python
#Quantas vezes cada área contaminada aparece nesses dados é uma análise que sempre fazemos na área ambiental
#Para saber precisamos contar, então usaremos a função value_counts()
classif_cont = dados_cont['Classifica'].value_counts() #para acessar uma coluna precisamos usar '[]'
classif_cont
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10,6))

classif_cont.plot(kind= 'bar')

plt.xlabel('Classificação')
plt.ylabel('Ocorrências')
plt.title('Contagem de ocorrências por contagem de áreas contaminadas')
plt.xticks(rotation=45, ha='right')

plt.show()
```
